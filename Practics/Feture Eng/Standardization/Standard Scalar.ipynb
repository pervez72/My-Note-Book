```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head()
```

```python
df=df.iloc[:,2:]
```

```python
df.sample(5)
```

```python
# trining Data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('Purchased',axis=1),df['Purchased'],test_size=.3,random_state=0)
```

```python
x_test.shape, x_train.shape
```

# Standrad Scalar

```python
# file import
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
# fit the scalar tarin set ,it will learn the parameter
scaler.fit(x_train)

# Transfrom train and test set
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
# note : learn train set but tranfrom train and test both set 
```

```python
# it Show numpi array this is a problem, so it convert data frame
x_test_scaled
```

```python
# Create Data frame
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_train.columns)
```

```python
x_train_scaled
```

```python
# Original value show 
np.round(x_train.describe(),1)
```

```python
# Scaleing Data
np.round(x_train_scaled.describe(),1)
```

```python
## both are parfectly work
```

# Effect Scalening

```python
# 2 plot create and show
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))

# now plot present 
ax1.scatter(x_train['Age'],x_train['EstimatedSalary'])
ax1.set_title("Befor Scaleing")

 
ax2.scatter(x_train_scaled['Age'],x_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaleing")
plt.show()
```

```python
# Another Plot Show

fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))

ax1.set_title('Befor Scaleing')
sns.kdeplot(x_train['Age'],ax=ax1)
sns.kdeplot(x_train['EstimatedSalary'],ax=ax1)

ax2.set_title('After Scaleing')
sns.kdeplot(x_train_scaled['Age'],ax=ax2)
sns.kdeplot(x_train_scaled['EstimatedSalary'],ax=ax2)
plt.show()

# Note : fig 2 show that, both columns is very nice comparable,good work for ml
# note :  fig 1 show that, both columns is very bad comparable,its dont good work ml
```
