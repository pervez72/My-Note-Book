```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df=df.drop(columns=['PassengerId','Name','Cabin','Ticket'])
```

```python
df.head()
```

```python
# Setp 1 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2,random_state=42)


```

```python
df.isnull().sum()
```

```python
x_train.head(5)
```

```python
# hundle missing value 
from sklearn.impute import SimpleImputer

si_age=SimpleImputer()
si_Embarked=SimpleImputer(strategy='most_frequent')

x_train_age=si_age.fit_transform(x_train[['Age']])
x_train_embarked=si_Embarked.fit_transform(x_train[['Embarked']])


x_test_age=si_age.fit_transform(x_test[['Age']])
x_test_embarked=si_Embarked.fit_transform(x_test[['Embarked']])
```

```python
# mean value is replace 
#x_train_age
#x_train_embarked
```

```python
# now Encoder code
from sklearn.preprocessing import OneHotEncoder

ohe_sex=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
ohe_embarked=OneHotEncoder(sparse_output=False,handle_unknown='ignore')


x_train_sex=ohe_sex.fit_transform(x_train[['Sex']])
x_train_embarked=ohe_embarked.fit_transform(x_train[['Embarked']])

x_test_sex=ohe_sex.fit_transform(x_test[['Sex']])
x_test_embarked=ohe_embarked.fit_transform(x_test[['Embarked']])
```

```python
# Ensure that the same columns are dropped and concatenated
x_train_rem = x_train.drop(columns=['Sex', 'Age', 'Embarked'])
x_test_rem = x_test.drop(columns=['Sex', 'Age', 'Embarked'])

# Check the shapes of the remaining columns
print(x_train_rem.shape)  # Should be (712, 4)
print(x_test_rem.shape)   # Should be (179, 4)

# Now concatenate the encoded and imputed columns
x_train_transformed = np.concatenate((x_train_rem, x_train_age, x_train_sex, x_train_embarked), axis=1)
x_test_transformed = np.concatenate((x_test_rem, x_test_age, x_test_sex, x_test_embarked), axis=1)

# Check the shapes of the transformed data
print(x_train_transformed.shape)  # Should be (712, 11)
print(x_test_transformed.shape)   # Should be (179, 11)

# Now fit the model
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_train_transformed)

# Predict on the test set
y_predict = clf.predict(x_test_transformed)
```
